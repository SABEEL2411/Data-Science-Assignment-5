```python
#import libraries
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the dataset
dataset = pd.read_csv("covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
dataset.shape
```

```python
#Delete the useless columns
df = dataset.drop(["Lat","Long"], axis = 1 , inplace = True)
```

```python
dataset.head()
```

```python
#aggregate the rows by country 
corona_dataset_aggregated = dataset.groupby("Country/Region").sum(numeric_only=True)
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
corona_dataset_aggregated.loc["China"].T.plot()
corona_dataset_aggregated.loc["India"].T.plot()
corona_dataset_aggregated.loc["Spain"].T.plot()
plt.legend()
```

```python
#calculate good measure
corona_dataset_aggregated.loc["China"][:3].T.plot()
```

```python
#calculate first derivative 
corona_dataset_aggregated.loc["China"].diff().T.plot()
```

```python
#Maximux infection rate 
corona_dataset_aggregated.loc["China"].diff().max()

```

```python
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rate = []

for c in countries :
    max_infection_rate.append(corona_dataset_aggregated.loc[c].diff().max())

corona_dataset_aggregated["Max_infection_rate"] = max_infection_rate
```

```python
corona_dataset_aggregated
```

```python
#Create new dataframe 
corona_data = pd.DataFrame(corona_dataset_aggregated["Max_infection_rate"])
```

```python
corona_data
```

```python
# importing happiness data set
happines_report=pd.read_csv("worldwide_happiness_report.csv")
happines_report
```

```python
# drop useless coloumn
useless_coloumn = ["Overall rank" , "Score", "Generosity", "Perceptions of corruption"]
happines_report.drop(useless_coloumn , axis = 1 , inplace = True )
happines_report
```

```python
happines_report.set_index("Country or region" , inplace = True)
happines_report.head()
```

```python
# Joining The Dataset 
corona_data.shape
```

```python
happines_report.shape
```

```python
data = corona_data.join(happines_report,how= "inner")
data
```

```python
data.corr()
```
